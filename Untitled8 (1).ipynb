```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# Ruta a tu ChromeDriver
driver_path = "ruta/al/chromedriver"  # Cambia esto por la ruta a tu ChromeDriver
service = Service(driver_path)

# Configurar el navegador
driver = webdriver.Chrome(service=service)

# URL de la pÃ¡gina con los datos de ingresos de GameStop
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
driver.get(url)

# Esperar unos segundos para que el contenido de la pÃ¡gina se cargue
time.sleep(5)

# Obtener el contenido HTML de la pÃ¡gina
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# Buscar la tabla de ingresos
tables = soup.find_all("table")
revenue_table = None

for table in tables:
    if "GameStop Quarterly Revenue" in str(table):  # Busca la tabla correcta
        revenue_table = table
        break

# Extraer los datos de la tabla
if revenue_table:
    rows = revenue_table.find_all("tr")
    revenue_data = []
    for row in rows:
        cols = row.find_all("td")
        revenue_data.append([col.text.strip() for col in cols])

    # Crear el DataFrame
    gme_revenue = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])

    # Limpiar los datos
    gme_revenue = gme_revenue.dropna()  # Eliminar filas vacÃ­as
    gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(",", "").str.replace("$", "")
    gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")

    # Mostrar las Ãºltimas cinco filas
    print(gme_revenue.tail())
else:
    print("No se pudo extraer la tabla de ingresos.")

# Cerrar el navegador
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la pÃ¡gina con los datos
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Cabeceras para simular un navegador real
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

# Realizar la solicitud
response = requests.get(url, headers=headers)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Buscar la tabla de ingresos
    tables = soup.find_all("table")
    revenue_table = None

    for table in tables:
        if "GameStop Quarterly Revenue" in str(table):  # Encuentra la tabla correcta
            revenue_table = table
            break

    # Extraer los datos de la tabla
    if revenue_table:
        rows = revenue_table.find_all("tr")
        revenue_data = []
        for row in rows:
            cols = row.find_all("td")
            revenue_data.append([col.text.strip() for col in cols])

        # Crear el DataFrame
        gme_revenue = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])

        # Limpiar los datos
        gme_revenue = gme_revenue.dropna()  # Eliminar filas vacÃ­as
        gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(",", "").str.replace("$", "")
        gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")

        # Mostrar las Ãºltimas cinco filas
        print(gme_revenue.tail())
    else:
        print("No se pudo encontrar la tabla de ingresos.")
else:
    print(f"Error al acceder a la pÃ¡gina: {response.status_code}")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar datos de las acciones de Tesla
tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-12-31")

# FunciÃ³n para graficar los datos
def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', linewidth=2)
    plt.title(title, fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Stock Price (USD)", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)
    plt.show()

# Llamar a la funciÃ³n para graficar
make_graph(tesla_data, "Tesla Stock Price (2020-2023)")
```
